```python
# Calculate a mapping: {item_id: target_mean}
item_id_target_mean = data.groupby('Product_Fat_Content').Product_Supermarket_Sales.mean()

# In our non-regularized case we just *map* the computed means to the `item_id`'s
data['item_target_enc'] = data['Product_Fat_Content'].map(item_id_target_mean)

# Fill NaNs

# Print correlation
encoded_feature = data['item_target_enc'].values
print(np.corrcoef(data['Product_Supermarket_Sales'].values, encoded_feature)[0][1])
```

```python
'''
     Differently to `.target.mean()` function `transform` 
   will return a dataframe with an index like in `all_data`.
   Basically this single line of code is equivalent to the first two lines from of Method 1.
'''
data['item_targete'] = data.groupby('Product_Fat_Content')['Product_Supermarket_Sales'].transform('mean')

# Fill NaNs

# Print correlation
encoded_feature = data['item_targete'].values
print(np.corrcoef(data['Product_Supermarket_Sales'].values, encoded_feature)[0][1])
```

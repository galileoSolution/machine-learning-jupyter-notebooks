```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test = pd.read_csv('test_fresh.xls')
```

```python
test.drop(['Product_Supermarket_Identifier','Product_Identifier','Supermarket_Identifier'], axis =1, inplace = True)
```

```python
train_test_split = [test]
supermarket_size = {'High': 2,'Medium': 1,'Small':0}
for dataset in train_test_split:
    dataset['Supermarket _Size'] =dataset['Supermarket _Size'].map(supermarket_size)
```

```python
test.isnull().sum()
```

```python
test['Product_Weight'].fillna(test.groupby('Product_Fat_Content')['Product_Weight'].transform('mean'), inplace = True)
```

```python
test['Supermarket _Size'].fillna(test.groupby('Product_Type')['Supermarket _Size'].transform('median'), inplace = True)
```

```python
dum1 = pd.get_dummies(test['Product_Fat_Content'])
dum2 = pd.get_dummies(test['Supermarket_Location_Type'])
dum3 = pd.get_dummies(test['Supermarket_Type'])
dum4 = pd.get_dummies(test['Product_Type'])
```

```python
test = pd.concat([test,dum1,dum2,dum3,dum4],axis = 1)
```

```python
test.drop(['Product_Fat_Content','Product_Type','Supermarket_Location_Type','Supermarket_Type'], axis =1 , inplace = True)
```

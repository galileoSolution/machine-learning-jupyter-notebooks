```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('best_fit.xls')
```

```python
data.drop(['Unnamed: 0','sss'],axis = 1,inplace = True)
```

```python
data.head()
```

```python
data['fee'] = data['Product_Price'] - data['Product_Weight']
```

```python
data.head()
```

```python
ss = data.Product_Price.apply(np.ceil)
```

```python
mm = data.Product_Weight.apply(np.ceil)
```

```python
ss = pd.DataFrame({'ss': ss})
```

```python
mm = pd.DataFrame({'mm': mm})
```

```python
data = pd.concat([data,ss,mm], axis = 1)
```

```python
data.head()
```

```python
data['first'] = data['mm'] - data['Product_Weight']
```

```python
data['second'] = data['ss'] - data['Product_Price']
```

```python
data.head()
```

```python
data.drop(['ss','mm'],axis = 1,inplace = True)
```

```python
data.head()
```

```python
X = data.drop('Product_Supermarket_Sales',axis =1)
```

```python
y = data['Product_Supermarket_Sales']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
predic = model.predict(X_test)
```

```python
from sklearn.cross_validation import cross_val_score
```

```python
scores = cross_val_score(model,X_test,y_test,cv =10)
```

```python
print(scores.mean()*100)
```

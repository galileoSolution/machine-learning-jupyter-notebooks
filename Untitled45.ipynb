```python
dum1 = pd.get_dummies(data['Gender'])
dum2 = pd.get_dummies(data['Location'])
dum3 = pd.get_dummies(data['famsize'])
dum4 = pd.get_dummies(data['Pstatus'])
dum5 = pd.get_dummies(data['schoolsup'])
dum6 = pd.get_dummies(data['paid'])
dum7 = pd.get_dummies(data['activities'])
dum8 = pd.get_dummies(data['nursery'])
dum9 = pd.get_dummies(data['higher'])
dum10 = pd.get_dummies(data['internet'])
dum11 = pd.get_dummies(data['famsup'])
```

```python
data = pd.concat([data,dum1,dum2,dum3,dum4,dum5,dum6,dum7,dum8,dum9,dum10,dum11], axis =  1)
```

```python
data.head()
```

```python
data.drop(['Gender','Location', 'famsize', 'Pstatus', 'schoolsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'famsup'], axis = 1 ,inplace = True)
```

```python
X = data.drop('Score', axis = 1)
y = data['Score']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

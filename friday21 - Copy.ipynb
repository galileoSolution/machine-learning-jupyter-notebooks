```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('angry.xls')
```

```python
data = data1
```

```python
data1.head()
```

```python
data1.drop(['Unnamed: 0','Product_Shelf_Visibility','Product_Type'],axis = 1,inplace = True)
```

```python
data1.head()
```

```python
data1.isnull().sum()
```

```python
dum1 = pd.get_dummies(data1['Product_Fat_Content'])
```

```python
dum2 = pd.get_dummies(data1['Supermarket_Location_Type'])
```

```python
dun3 = pd.get_dummies(data1['Supermarket_Type'])
```

```python
data = pd.concat([data1,dum1,dum2,dun3],axis =1)
```

```python
data.head()
```

```python
data.drop('Supermarket _Size',axis = 1,inplace = True)
```

```python
data.head()
```

```python
data.drop(['Product_Fat_Content','Supermarket_Location_Type','Supermarket_Type'],axis = 1 ,inplace = True)
```

```python
data.head()
```

```python
data['f'] = data.Product_Weight.astype(str)
```

```python
data.head()
```

```python
data['f'] = data['f'].apply(lambda x: x.split('.')[0])
```

```python
data.head()
```

```python
data['f'] = data['f'].astype(float)
```

```python
data.dtypes
```

```python
data['f'] =  data['Product_Weight'] - data['f']
```

```python
data.head()
```

```python
data['Product_Weight'].nunique()
```

```python
X = data.drop('Product_Supermarket_Sales',axis =1)
```

```python
y = data['Product_Supermarket_Sales']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
plt.scatter(y_test,prediction)
```

```python
from sklearn.cross_validation import cross_val_score
scored = cross_val_score(model,X_test,y_test,cv =10)
```

```python
print(scored.mean()*100)
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
param_grid = {'gamma': [1,0.1,0.01,0.001,0.0001], 'max_depth': [3, 4, 5]}
```

```python
grid = GridSearchCV(XGBRegressor(),param_grid,verbose =3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
griedp = grid.predict(X_test)
```

```python
plt.scatter(y_test,griedp)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('train_fresh.xls')
```

```python
data.drop(['Product_Supermarket_Identifier','Product_Identifier','Supermarket_Identifier'], axis =1, inplace = True)
```

```python
data.head()
```

```python
train_test_split = [data]
supermarket_size = {'High': 2,'Medium': 1,'Small':0}
for dataset in train_test_split:
    dataset['Supermarket _Size'] =dataset['Supermarket _Size'].map(supermarket_size)
```

```python
data.head()
```

```python
data['Supermarket _Size'].fillna(data.groupby('Product_Type')['Supermarket _Size'].transform('median'), inplace = True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Product_Weight'].fillna(data.groupby('Product_Fat_Content')['Product_Weight'].transform('mean'), inplace = True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.head()
```

```python
data.to_csv('angry.xls')
```

```python
dum1 = pd.get_dummies(data['Product_Fat_Content'], drop_first= True)
```

```python
dum2 = pd.get_dummies(data['Supermarket_Location_Type'], drop_first= True)
```

```python
dum3 = pd.get_dummies(data['Supermarket_Type'], drop_first= True)
```

```python
dum4 = pd.get_dummies(data['Product_Type'])
```

```python
data.groupby('Product_Type').size()
```

```python
dum4.columns
```

```python
dum4.drop(['Baking Goods', 'Breads', 'Breakfast', 'Canned', 'Dairy', 'Hard Drinks','Health and Hygiene', 'Meat', 'Others', 'Seafood','Soft Drinks', 'Starchy Foods'])
```

```python
dum4
```

```python
data = pd.concat([data,dum1,dum2,dum3,dum4],axis = 1)
```

```python
data.head()
```

```python
data.drop(['Product_Fat_Content','Product_Type','Supermarket_Location_Type','Supermarket_Type'], axis =1 , inplace = True)
```

```python
data.head()
```

```python
data.columns
```

```python
X = data[['Product_Weight', 'Product_Shelf_Visibility', 'Product_Price',
       'Supermarket_Opening_Year', 'Supermarket _Size',
        'Normal Fat', 'Ultra Low fat', 'Cluster 2',
       'Cluster 3', 'Supermarket Type1', 'Supermarket Type2',
       'Supermarket Type3', 'Frozen Foods', 'Fruits and Vegetables',
       'Household', 'Snack Foods']]
```

```python
y = data['Product_Supermarket_Sales']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
predi = model.predict(X_test)
```

```python
from sklearn.cross_validation import cross_val_score
```

```python
scores = cross_val_score(model,X_test,y_test,cv =10)
```

```python
print(scores.mean()*100)
```

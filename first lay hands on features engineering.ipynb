```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['title'] = df['Name'].str.extract(r',\s*([^\.]*)\s*\.', expand=False)
```

```python
df['title'].value_counts()
```

```python
df.head()
```

```python
df['title'].unique()
```

```python
title_mapping = {'Mr': 0, 'Mrs':1, 'Miss':2, 'Master': 3, 'Don': 3, 'Rev':3, 'Dr':3, 'Mme':3, 'Ms':3,
       'Major':3, 'Lady':3, 'Sir':3, 'Mlle':3, 'Col':3, 'Capt':3, 'the Countess':3,
       'Jonkheer':3}
df['title'] = df['title'].map(title_mapping)
```
